```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Intel/fid_flan_t5_base_nq")
model = AutoModelForSeq2SeqLM.from_pretrained("Intel/fid_flan_t5_base_nq")
```

## Generate Research Book Name

```python
input_text = "Give 3 Short, Unqiue and Good Legal Research title names on the legal query - 'Siblings Murder'"
input_ids = tokenizer(input_text, return_tensors="pt").input_ids

outputs = model.generate(input_ids, max_length=100)
print(tokenizer.decode(outputs[0]))
```

## Enhance Facts

```python
fact='Loan pending with the bank'
prompt = f"Enhanced legal version of the fact -'{fact}' with the addition of 2-5 words."
input_ids = tokenizer(prompt, return_tensors="pt").input_ids

outputs = model.generate(input_ids, max_length=100)
print(tokenizer.decode(outputs[0]))
```
