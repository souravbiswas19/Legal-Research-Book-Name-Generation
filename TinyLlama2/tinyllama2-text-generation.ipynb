```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("TinyLlama/TinyLlama-1.1B-Chat-v1.0")
model = AutoModelForCausalLM.from_pretrained("TinyLlama/TinyLlama-1.1B-Chat-v1.0")
```

## Generate Research Book Name

```python
query="Siblings Murder"
```

```python
def generate_titles(query):
    prompt = f"Generate 3 Short, Unqiue and Good Legal Research title names on the legal query -'{query}'"
    inputs = tokenizer(prompt, return_tensors="pt", return_attention_mask=False)
    outputs = model.generate(**inputs, max_length=100)
    return tokenizer.batch_decode(outputs)[0]
```

```python
print(generate_titles(query))
```

## Enhance Facts

```python
fact='Loan pending with the bank'
prompt = f"Enhance the fact -'{fact}' legally by adding 2-3 words"
inputs = tokenizer(prompt, return_tensors="pt", return_attention_mask=False)
outputs = model.generate(**inputs, max_length=100)
tokenizer.batch_decode(outputs)[0]
```
